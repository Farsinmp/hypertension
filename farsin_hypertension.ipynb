# **Introduction**
The dataset hypertension_data.csv is a collection of information designed to study hypertension (high blood pressure) and its associated risk factors. Hypertension is a critical public health concern, often referred to as the "silent killer" because it typically presents no symptoms until severe complications arise. This dataset provides valuable insights into various demographic, health, and lifestyle variables that influence the onset and progression of hypertension.

**Key Features of the Dataset:**
Demographic Information:
Includes age, gender, and possibly geographic or socio-economic data. These variables help analyze how hypertension prevalence varies across different population segments.

**Health Indicators:**
Variables such as blood pressure readings (systolic and diastolic), BMI (Body Mass Index), cholesterol levels, and glucose levels provide quantitative measures of health status.

**Lifestyle Factors:**
Data on smoking, alcohol consumption, physical activity, and dietary habits that are known contributors to hypertension.

**Medical History:**
Includes pre-existing conditions such as diabetes, cardiovascular diseases, or kidney issues that might correlate with hypertension risk.

**Target Variable:**
The dataset likely includes a target variable indicating whether an individual has hypertension (binary: hypertensive vs. non-hypertensive or a range of severity levels).

# **Outlier** **Detection**

```python
plt.figure(figsize=(12,8))
sns.boxplot(df, orient='h')
plt.show()
```

```python
Q1=np.quantile(df,0.25)
Q2=np.quantile(df,0.50)
Q3=np.quantile(df,0.75)
Q1,Q2,Q3
IQR=Q3-Q1
IQR
```

```python
def remove_outliers_iqr(df, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile) for the column
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Define the lower and upper bounds for detecting outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out the outliers from the DataFrame
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered


df_no_outliers = remove_outliers_iqr(df,["oldpeak","ca","thalach"])



plt.figure(figsize=(10,10))
sns.boxplot(df,orient='h',palette='Set1')
plt.show()
```
